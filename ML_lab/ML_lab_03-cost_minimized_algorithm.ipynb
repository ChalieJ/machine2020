# Hypothesis

Linear Regression 을 구하기 위해 1차 방정식으로 가설을 세우는 것.

간단하게 설명하기 위해 아래와 같이 식을 만든다

H(x) = Wx

이 Hypothesis에 아래 값을 적용한다.

|x|y|
|-|-|
|1|1|
|2|2|
|3|3|

위의 Hypothesis와 x y 값을 가지고 cost function 실제 값을 확인해본다.

cost(W)= $ \frac{1}{m}\sum_{i=1}^{m}(Wx^{(i)}-y^{(i)})^2 $

W = 1, cost(W) 값을 계산해보자 \
$\frac{1}{3}$((1 * 1 - 1)$^2$ + (1 * 2 - 2)$^2$ + (1 * 3 - 3)$^2$) = 0

W = 0, cost(W) 값을 계산해 보자 \
$\frac{1}{3}$((0 * 1 - 1)$^2$ + (0 * 2 - 2)$^2$ + (0 * 3 - 3)$^2$) = 14 / 3 = 4.67

W = 2, cost(W) 값을 계산해 보자 \
$\frac{1}{3}$((2 * 1 - 1)$^2$ + (2 * 2 - 2)$^2$ + (2 * 3 - 3)$^2$) = 14 / 3 = 4.67

W=1, cost(W)=0
W=0, cost(W)=4.67
W=2, cost(W)=4.67

대략 이차방정식 느낌의 그래프가 나타난다.
여기서 코스트가 최소화 되는 점을 찾는다. (기계적으로)

# Gradient Descent Algorithm

경사를 내려가는 알고리즘 \
cost function을 minimize하는데 사용됨

경사를 따라 내려가다가, 경사가 없어지는 지점에서 멈춘다 \
- 아무점에서 시작해도 된다
- Weight값을 변경하여, 코스트가 줄어드는지 확인. 반복
- 경사는 미분을 사용하여 구한다

# 중간 정리

## 1. Linear regrration
-> 여러가지 좌표값 [x, y]들을, 1차 함수(좌표상의 직선)와 평균값이 가장 비슷한 값을 찾아내는것




## 2. Hypothesis
-> 주어진 값을 어떻게 예측할것인가에 대한 식 (대략 1차 함수, 즉 좌표상의 직선)

H(x) = Wx + b

W : Weight (기울기)
b : bias (0점으로 부터 y축 offset)

## 3. cost function
-> Hypothesis로 예측한 값이 얼마나 정확한지 계산하는 방법 \
-> x값 (feature, 특성)을 사용하여, Hypothesis로 계산한 값과 y의 값의 차이를 모두 계산하여, 평균으로 나타낸 값.



## 4. cost function minimize - Gradient Descent Algorithm
-> 경사면을 내려가면서 기울기가 0인 지점을 찾는 알고리즘 \
-> 경사도는 미분을 통해 구한다 \
-> 기울기가 - 이면 Weight을 +값을 추가 해준다. 반대로 기울기가 + 이면 Weight을 -값을 추가해준다.
